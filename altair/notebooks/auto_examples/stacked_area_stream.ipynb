# Altair Example: Stacked Area Stream

--



## Load Dataset
The data comes in the form of a Pandas Dataframe:

```python
from altair import load_dataset
data = load_dataset("unemployment-across-industries")
data.head()
```

## Define Altair Specification

```python
from altair import *  # Import the altair API

chart = Chart(data).mark_area(
    stacked='center',
).encode(
    color=Color('series:N',
        scale=Scale(
            range='category20b',
        ),
    ),
    x=X('date:T',
        axis=Axis(
            axisWidth=0.0,
            format='%Y',
            labelAngle=0.0,
            tickSize=0.0,
        ),
        scale=Scale(
            nice='month',
        ),
        timeUnit='yearmonth',
    ),
    y=Y('sum(count):Q',
        axis=False,
    ),
).configure_cell(
    height=200.0,
    width=300.0,
)
```
