# Basic IPython Widget for Altair

This notebook shows how to create a simple user interface for working with an Altair `Viz` object. This version uses IPython's interactive widgets (`ipywidgets`).

## Setup

```python
import seaborn.apionly as sns
import altair.api as alt
import pandas as pd
```

## Load data

```python
df = sns.load_dataset('titanic')
```

```python
df = df.dropna()
```

```python
df.head()
```

# Widget view

The `view` submodule in Altair has an `AltairWidget` that will return an IPython `Widget` that is bound to the Altair objects. As you change the UI controls, the Altair objects are updated dynamically. Currently, this doesn't work with any of our renderers because of bugs in the renderers. Instead, a text box is shown.

```python
from altair import view
```

```python
v1 = alt.Viz(df).encode(x='age', y='fare').point()
```

```python
aw = view.AltairWidget(v1, shelves=('x','y','color','size'))
aw.render()
```
