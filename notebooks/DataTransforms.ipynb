```python
def bin_Q(series, maxbins):
    bin_series, bins = pd.cut(series, maxbins, include_lowest=True, retbins=True, labels=False)
    return pd.Series(bins[bin_series])
```

```python
df['bin_amount'] = bin_Q(df.amount, 5)
```
