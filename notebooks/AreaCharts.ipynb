# Area Charts

```python
from altair import *
```

```python
data = load_dataset('unemployment-across-industries')
data['date'] = data['date'].astype(str)
```

## Simple area chart

```python
Chart(data).mark_area().encode(
    X('date:T', axis=Axis(format='%Y', labelAngle=0)),
    y='sum(count)'
)
```

## Stacked area chart

```python
Chart(data).mark_area().encode(
    X('date:T',
      timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    Color('series:N', scale=Scale(range="category20b"))
)
```

## Normalized area chart

```python
Chart(data).mark_area(stacked='normalize').encode(
    X('date:T', timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    color=Color('series:N', scale=Scale(range="category20b"))
)
```

## Streamgraph

```python
Chart(data).mark_area(stacked='center').encode(
    X('date:T', timeUnit='yearmonth',
      axis=Axis(format='%Y', labelAngle=0, title='year')),
    Y('sum(count)'),
    color=Color('series:N', scale=Scale(range="category20b"))
)
```
