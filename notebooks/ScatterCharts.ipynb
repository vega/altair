# Scatter-Plot Examples

```python
from altair import *
```

```python
cars = load_dataset('cars')
```

## Dot Plot

```python
Chart(cars).mark_point().encode(
    x='Horsepower',
)
```

## Scatter Plot

```python
Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon'
)
```

## Bubble Plot

```python
Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    size='Acceleration'
)
```

## Scatter Plot with Color and/or Shape

```python
Chart(cars).mark_point().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    shape='Origin',
)
```

## Circle Scatterplot

```python
Chart(cars).mark_circle().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

## Square ScatterPlot

```python
Chart(cars).mark_square().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
)
```

## Text ScatterPlot

```python
Chart(cars).mark_text().encode(
    x='Horsepower',
    y='Miles_per_Gallon',
    color='Origin',
    text='Origin'
)
```

## Tick Plot

```python
Chart(cars).mark_tick().encode(
    x='Horsepower:Q',
    y='Cylinders:O',
)
```
